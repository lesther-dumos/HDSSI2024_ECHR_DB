```python
import pandas as pd
import os
import json

# Load data
population_df = pd.read_csv('Data/Population/API_SP.POP.TOTL_DS2_en_csv_v2_739890 - API_SP.POP.TOTL_DS2_en_csv_v2_739890.csv')
echrdb_df1 = pd.read_excel('Output/business_up_to_2022.xlsx')

# Combine GeoJSON files
json_files = ["belgium.json", "cyprus.json", "austria.json", "switzerland.json", "germany.json",
              "denmark.json", "spain.json", "france.json", "greece.json", "italy.json", "ireland.json",
              "iceland.json", "malta.json", "norway.json", "netherlands.json", "portugla.json", "sweden.json",
              "findland.json", "turkey.json", "united_kingdom.json", "albania.json", "andorra.json", 
              "armenia.json", "azerbaijan.json", "bulgaria.json", "croatia.json", "estonia.json", "georgia.json",
              "hungary.json", "latvia.json", "liechtenstein.json", "lithuania.json", "moldova.json", "poland.json",
              "romania.json", "russia.json", "san_marino.json", "slovakia.json", "slovenia.json", "ukraine.json",
              "monaco.json", "montenegro.json", "macedonia.json", "bosnia_herzegovina.json", "luxemburg.json",
              "czech_republic.json", "serbia.json"]

data_dir = "country_shp"
all_features = []

for json_file in json_files:
    file_path = os.path.join(data_dir, json_file)
    with open(file_path, "r") as f:
        data = json.load(f)
        if 'features' in data:
            all_features.extend(data['features'])

combined_geojson = {
    "type": "FeatureCollection",
    "features": all_features
}

output_dir = "output"
output_file = os.path.join(output_dir, "combined3.json")

os.makedirs(output_dir, exist_ok=True)

with open(output_file, "w") as f:
    json.dump(combined_geojson, f, indent=4)

print(f"Combined GeoJSON saved to {output_file}")

# Map country codes to names
country_dict = {
    1: 'Austria', 2: 'Belgium', 3: 'Switzerland', 4: 'Cyprus', 5: 'Germany', 6: 'Denmark', 7: 'Spain',
    8: 'France', 9: 'Greece', 10: 'Italy', 11: 'Ireland', 12: 'Iceland', 13: 'Luxembourg', 14: 'Malta',
    15: 'Norway', 16: 'Netherlands', 17: 'Portugal', 18: 'Sweden', 19: 'Finland', 20: 'Turkey', 21: 'United Kingdom',
    22: 'Albania', 23: 'Andorra', 24: 'Armenia', 25: 'Azerbaijan', 26: 'Bulgaria', 27: 'Croatia', 28: 'Czech Republic',
    29: 'Estonia', 30: 'Georgia', 31: 'Hungary', 32: 'Latvia', 33: 'Liechtenstein', 34: 'Lithuania', 35: 'Moldova',
    36: 'Poland', 37: 'Romania', 38: 'Russia', 39: 'San Marino', 40: 'Slovakia', 41: 'Slovenia', 42: 'North Macedonia',
    43: 'Ukraine', 44: 'Bosnia Herzegovina', 46: 'Monaco', 47: 'Montenegro', 48: 'Serbia'
}

echrdb_df1['country1'] = echrdb_df1['country1'].map(country_dict)

# Define countries list
countries = list(country_dict.values())
countries.extend(['Russian Federation', 'Bosnia and Herzegovina', 'Czechia', 'Turkiye', 'Slovak Republic'])

# Filter population data for specified countries
filtered_population_df = population_df[population_df['Country Name'].isin(countries)]

# Adjust country names in the population DataFrame
replace_dict = {
    'Turkiye': 'Turkey',
    'Russian Federation': 'Russia',
    'Bosnia and Herzegovina': 'Bosnia Herzegovina',
    'Czechia': 'Czech Republic',
    'Slovak Republic': 'Slovakia'
}

filtered_population_df['Country Name'] = filtered_population_df['Country Name'].replace(replace_dict)

# Drop columns for years not needed
years_to_drop = [str(year) for year in range(1960, 1970)] + ['2023']
filtered_population_df.drop(columns=years_to_drop, inplace=True)

print(len(filtered_population_df)) # 47
print(len(countries))              # 47

filtered_population_df.head()

# Filter ECHR database for years >= 1970
echrdb_df1 = echrdb_df1[echrdb_df1['dtelgd'] >= 1970]

# Define years of interest
years_interested = list(range(1970, 2023))

# Initialize results dictionary
results_by_year = {year: {country: {'no_violation': 0, 'violation': 0} for country in country_dict.values()} for year in years_interested}

# Loop through each year and country to count violations
for year in years_interested:
    for country in country_dict.values():
        country_df1 = echrdb_df1[(echrdb_df1['country1'] == country) & (echrdb_df1['dtejmt'] <= year)]
        
        for col in country_df1.iloc[:, 19:37]:
            no_violations = (country_df1[col] == 1).sum()
            violations = (country_df1[col] == 2).sum()
            
            results_by_year[year][country]['no_violation'] += no_violations
            results_by_year[year][country]['violation'] += violations

# Convert results into a DataFrame
results_list = []
for year in years_interested:
    for country, data in results_by_year[year].items():
        row = {
            'country': country,
            'year': year,
            'no_violation': data['no_violation'],
            'violation': data['violation'],
            'total': data['no_violation'] + data['violation'],
        }
        results_list.append(row)

countries_violations_by_year = pd.DataFrame(results_list)

# Melt population data for merging, excluding non-year columns
filtered_population_df_melted = filtered_population_df.melt(id_vars=['Country Name'], var_name='year', value_name='population')
filtered_population_df_melted = filtered_population_df_melted[filtered_population_df_melted['year'].str.isnumeric()]
filtered_population_df_melted.columns = ['country', 'year', 'population']
filtered_population_df_melted['year'] = filtered_population_df_melted['year'].astype(int)

# Merge violations and population data
countries_violations_by_year = countries_violations_by_year.merge(filtered_population_df_melted, on=['country', 'year'], how='left')

print(countries_violations_by_year)
```

```python
import pandas as pd
import os
import json

# Assuming your previous code has been run and `countries_violations_by_year` DataFrame is available

# Define the directory and json file names
data_dir = "country_shp"
json_files = ["belgium.json", "cyprus.json", "austria.json", "switzerland.json", "germany.json",
              "denmark.json", "spain.json", "france.json", "greece.json", "italy.json", "ireland.json",
              "iceland.json", "malta.json", "norway.json", "netherlands.json", "portugla.json", "sweden.json",
              "findland.json", "turkey.json", "united_kingdom.json", "albania.json", "andorra.json", 
              "armenia.json", "azerbaijan.json", "bulgaria.json", "croatia.json", "estonia.json", "georgia.json",
              "hungary.json", "latvia.json", "liechtenstein.json", "lithuania.json", "moldova.json", "poland.json",
              "romania.json", "russia.json", "san_marino.json", "slovakia.json", "slovenia.json", "ukraine.json",
              "monaco.json", "montenegro.json", "macedonia.json", "bosnia_herzegovina.json", "luxemburg.json",
              "czech_republic.json", "serbia.json"]

# Load all GeoJSON files
all_features = []
for json_file in json_files:
    file_path = os.path.join(data_dir, json_file)
    with open(file_path, "r") as f:
        data = json.load(f)
        if 'features' in data:
            all_features.extend(data['features'])

# Add violation data to GeoJSON properties
for feature in all_features:
    country_name = feature['properties'].get('COUNTRY')
    # Check if the country_name is in the DataFrame index
    if country_name in countries_violations_by_year['country'].values:
        # Filter the DataFrame for the specific country
        country_data = countries_violations_by_year[countries_violations_by_year['country'] == country_name]
        for _, row in country_data.iterrows():
            year = row['year']
            feature['properties'][f'no_violation_{year}'] = int(row['no_violation']) if pd.notna(row['no_violation']) else None
            feature['properties'][f'violation_{year}'] = int(row['violation']) if pd.notna(row['violation']) else None
            feature['properties'][f'total_{year}'] = int(row['total']) if pd.notna(row['total']) else None
            feature['properties'][f'population_{year}'] = int(row['population']) if pd.notna(row['population']) else None

# Create the combined GeoJSON
combined_geojson = {
    "type": "FeatureCollection",
    "features": all_features
}

# Define the output directory and file name
output_dir = "output"
output_file = os.path.join(output_dir, "combined_violations_population.json")

# Ensure the output directory exists
os.makedirs(output_dir, exist_ok=True)

# Write the combined GeoJSON to a file
with open(output_file, "w") as f:
    json.dump(combined_geojson, f, indent=4)

print(f"Combined GeoJSON with violations and population data saved to {output_file}")
```
