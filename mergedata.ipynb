```python
import pandas as pd
import numpy as np

pd.options.display.max_columns = None
pd.options.display.max_rows = None
pd.set_option('display.max_columns', None)
```

```python
# Business Files
bus_comp = pd.read_csv('Data/Bus Data/ECHRdb_Comprehensive_BUS_PB&LD.xlsx - BUS.csv')
bus_2021 = pd.read_csv('Data/Bus Data/ECHRdb_2021-2022_BUS_PB&LD.xlsx - Sheet1.csv')
```

```python
len(bus_comp) # 21582
len(bus_2021) # 2227


merged_bus = pd.concat([bus_comp, bus_2021], ignore_index=True)

len(merged_bus) # 24790
merged_bus.head()

merged_bus.fillna(0, inplace=True)

```
