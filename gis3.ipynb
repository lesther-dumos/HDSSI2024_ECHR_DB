```python
country_pop_melted = country_pop.melt(id_vars=['Country'], 
                                      var_name='Year', 
                                      value_name='Population')

country_pop_melted['Year'] = country_pop_melted['Year'].astype(int)
country_pop_melted['Population'] = country_pop_melted['Population'].astype(int)
```

```python
import pandas as pd

# Assume echrdb_df is already defined
# Years of interest
years_interested = [1970, 1980, 1990, 2000, 2010]

# Initialize the results dictionary
results_by_year = {year: {country: {'no_violation': 0, 'violation': 0} for country in country_dict.values()} for year in years_interested}

# Loop through each year
for year in years_interested:
    for country in country_dict.values():
        country_df1 = echrdb_df[(echrdb_df['country1'] == country) & (echrdb_df['dtejmt'] <= year)]
        
        for col in country_df1.iloc[:, 19:37]:
            no_violations = (country_df1[col] == 1).sum()
            violations = (country_df1[col] == 2).sum()
            
            results_by_year[year][country]['no_violation'] += no_violations
            results_by_year[year][country]['violation'] += violations

# Convert the results into a DataFrame
results_list = []
for year in years_interested:
    for country, data in results_by_year[year].items():
        row = {
            'country': country,
            'year': year,
            'no_violation': data['no_violation'],
            'violation': data['violation'],
            'total': data['no_violation'] + data['violation'],
        }
        results_list.append(row)

countries_violations_by_year = pd.DataFrame(results_list)

# Add population data
country_pop_melted.columns = ['country', 'year', 'population']
countries_violations_by_year = countries_violations_by_year.merge(country_pop_melted, on=['country', 'year'], how='left')

print(countries_violations_by_year)

print(countries_violations_by_year)
```

```python
import os
import json
import pandas as pd

# Ensure 'countries_violations_by_year' and 'country_pop_melted' dataframes are already defined
# Ensure 'json_files' is a list of your GeoJSON filenames

# Environment
data_dir = "country_shp"
all_features = []

# Load all features from the shapefiles
for json_file in json_files:
    file_path = os.path.join(data_dir, json_file)
    with open(file_path, "r") as f:
        data = json.load(f)
        if 'features' in data:
            all_features.extend(data['features'])

# Add violation and population data to GeoJSON properties
new_features = []
for feature in all_features:
    country_name = feature['properties'].get('COUNTRY')
    
    # Filter data for the current country
    country_data = countries_violations_by_year[countries_violations_by_year['country'] == country_name]
    
    if not country_data.empty:
        for _, row in country_data.iterrows():
            new_feature = feature.copy()
            year = row['year']
            no_violation = row['no_violation']
            violation = row['violation']
            total = row['total']
            population = row['population']
            
            new_feature['properties'] = new_feature.get('properties', {}).copy()  # Ensure we're copying properties correctly
            new_feature['properties']['year'] = year
            new_feature['properties']['no_violation'] = int(no_violation) if pd.notna(no_violation) else None
            new_feature['properties']['violation'] = int(violation) if pd.notna(violation) else None
            new_feature['properties']['total'] = int(total) if pd.notna(total) else None
            new_feature['properties']['population'] = int(population) if pd.notna(population) else None
            
            new_features.append(new_feature)
            print(f"Added feature for {country_name} in year {year}")
    else:
        print(f"No data found for {country_name}, keeping original feature.")
        new_features.append(feature)

combined_geojson = {
    "type": "FeatureCollection",
    "features": new_features
}

output_dir = "output"
output_file = os.path.join(output_dir, "gis_timelapse.json")

os.makedirs(output_dir, exist_ok=True)

with open(output_file, "w") as f:
    json.dump(combined_geojson, f, indent=4)

print(f"Combined GeoJSON Environment saved to {output_file}")
```
